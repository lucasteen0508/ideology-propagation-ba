```python
import pandas as pd
import numpy as np
import re

# Hydration Output laden
df = pd.read_csv("/Users/lucasteen/Desktop/ideology-classifier/ideology_classifier/get_seed_users/data/hydration_output.csv", dtype= str)

# Entferne alle leeren Spalten
df = df.dropna(how="all")
print(len(df))

# Alle Zeilen entfernen, wo Profiltext NaN ist
df = df.dropna(subset=["profile_text"])
print(len(df))

# Zeilen, in denen Profiltext nur ein leerer String ist checken -> Perfekt, für jeden User existiert ein Profil
empty_profile_text = df[df["profile_text"] == ""]
print(empty_profile_text)

# Reduktion auf einzigartige User -> Knapp 5k einzigartige User
df = df.drop_duplicates(subset = 'user_id')

# Filtere Datensatz nach Profiltexten, die keine Hashtags verwenden
no_hashtag_mask = ~df['profile_text'].str.contains(r"#\w+", na=False)

# Alle User ohne Hashtags im Profil-Text-> Auf Keywords gehen
users_without_hashtag = df.loc[no_hashtag_mask]

# Sehr genau definiert, um Ambiguitäten zu vermeiden
keywords_dict = {
    1: [
        "Gender",
        "Genderwahn",
        # "Irrenhaus", Das war ein extrem starker Seed-User während der Datensichtung
        "Heimat",
        "Konservativ",
        "Maskulin",
        "Hetero",
        "Vaterland",
        "Leitkultur",
        "Tradition",
        "Patriot",
        "Deutschland",
        "Identität",
        "Islamkritik",
        "Nation",
        "Deutsch",
        "Remigration"
    ],
    0: [
        "Anarchismus",
        #"Liberal", entfernt, zu arbiträr
        "Feminismus",
        "Queer",
        "LGBTQ",
        "Diversität",
        "Inklusion",
        #"Sozial", entfernt, zu schwach
        "Menschenrechte",
        "Antifa",
        "Gleichstellung",
        "RefugeesWelcome",
        "Rassismuskritik",
        "Nachhaltigkeit",
        "Klimawende",
        "Solidarität",
        #"Gerechtigkeit", zu ambig
        "Community",
        "Grün"
    ]
}

# Liste aller Stichproben speichern
sample_list = []

for label, words in keywords_dict.items():
    # \b sorgt für exakte Wortgrenzen, re.escape schützt Sonderzeichen
    pattern = r"\b(" + "|".join(words) + r")\b"
    df_label = users_without_hashtag[users_without_hashtag["profile_text"].str.contains(pattern, case=False, na=False)].copy()
    df_label = df_label.sample(n=min(200, len(df_label)), random_state=42)
    df_label["label"] = label  # Label-Spalte hinzufügen
    sample_list.append(df_label)

# Kombinieren
combined_samples = pd.concat(sample_list, ignore_index=True)

# Exportieren
#combined_samples.to_csv("keywords_sampled_combined.csv", index=False, encoding="utf-8")
print("CSV erzeugt: keywords_sampled_combined.csv")
```
